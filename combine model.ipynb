```python
# load text data
Id = []
from keras.preprocessing import sequence
import numpy as np
import logging
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
maxlen = 300 # sequence 取到的长度，用于下面的padding
input_dir = './vectors/'
count = 0
x=[]
import os
# 下面是按照listdir的顺序load vectors，实际应该要按数据对应的html文件的标号来load
for file in os.listdir(input_dir):
    count += 1
    this_x = np.load(os.path.join(input_dir, file))
    this_x = this_x.tolist()
    x.append(this_x)
    Id.append(file)
    if count % 100 == 0:
        logging.info('{} has loaded...'.format(count))
print('load {} vectors done'.format(count))
textseq = sequence.pad_sequences(x, maxlen=maxlen)
```

```python
#load screenshot data
img = []
num = []
input_dir = './render/'
import glob
import imageio
import cv2

# crop screeshot that are too long
def cropData():
    #filename = 'render'+id.split('.')[0].split('page')[1]+'.png'
    count = 0
    for f in glob.glob('./render/*.png'):
        count += 1
        filename = f.split('/')[-1]
        #print(filename)
        im = cv2.imread(f)
        if im.shape[0]>im.shape[1]:
            im = im[:im.shape[1],:im.shape[1],:]
        res = cv2.resize(im, dsize=(224, 224), interpolation=cv2.INTER_CUBIC)
        img.append(res)
        num.append(filename)
        logging.info('{} has loaded...'.format(count))

cropData()

```

```python
import csv
import random

entity_data = []
with open("./dataset.csv") as f:
    csv_reader = csv.DictReader(f)
    for row in csv_reader:
        if row['is entity']=='0' or row['is entity']=='1':
            entity_data.append(row)
        if row['categories'] != "":
            category_data.append(row)

random.shuffle(entity_data)
```

```python
LocImg = dict()
LocTxt = dict()
for i in range(len(num)):
    LocImg[num[i]]=i
for i in range(len(Id)):
    LocTxt[Id[i].split('.')[0]+'.html'] = i
Etxt = []
Eimg = []
Elab = []
for row in entity_data:
    Etxt.append(textseq[LocTxt[row['html'].split('/')[-1]]])
    Eimg.append(img[LocImg[row['render'].split('/')[-1]]])
    lab = np.array([1.,0.])
    if(row['is entity']=='1'):
        lab =np.array([0.,1.])
    Elab.append(lab)
Etxt = np.array(Etxt)
Eimg = np.array(Eimg)
Elab = np.array(Elab)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

for i in range(1):
    image = Eimg[i]
    label = Elab[i]
    if(label[0]<.5):
        continue
    print("Image shape: ", image.shape)
    print("Label: ", label)
    image_ = np.squeeze(image) # you can give axis attribute if you wanna squeeze in specific dimension
    plt.imshow(image_)
    plt.show()
    
partition_rate = 0.9
cnt = int(len(Etxt)*partition_rate)
Etrain_txt = Etxt[:cnt]
Etrain_img = Eimg[:cnt]
Etrain_lab = Elab[:cnt]
Etest_txt = Etxt[cnt:]
Etest_img = Eimg[cnt:]
Etest_lab = Elab[cnt:]
```

```python
from tensorflow import keras
from tensorflow.keras import layers

import tensorflow_hub as hub

IMG_HEIGHT = 224
IMG_WIDTH = 224


num_categories = 10
num_words = 89547+1  # Size of vocabulary obtained when preprocessing text data
text_input = keras.Input(shape=(None,), name='html-text')  # Variable-length sequence of ints
encoder_input = keras.Input(shape=(224, 224, 3), name='img')

# Embed each word in the text into a 64-dimensional vector
text_features = layers.Embedding(num_words, 64)(text_input)
# Reduce sequence of embedded words in the title into a single 128-dimensional vector
text_features = layers.LSTM(128)(text_features)

# Use pretrained model for img feature extraction
feature_extractor_url = "https://tfhub.dev/google/tf2-preview/mobilenet_v2/feature_vector/2" #@param {type:"string"}
feature_extractor_layer = hub.KerasLayer(feature_extractor_url,
                                         input_shape=(IMG_HEIGHT,IMG_WIDTH,3))
feature_extractor_layer.trainable = False
img_features = feature_extractor_layer(encoder_input)

x = layers.concatenate([text_features, img_features])

entity_pred = layers.Dense(2, activation='softmax', name='is_entity')(x)
category_pred = layers.Dense(num_categories, activation='softmax', name='category')(x)
#model = keras.Model(inputs=[text_input, encoder_input],
#                    outputs=[entity_pred, category_pred])
Entity_model = keras.Model(inputs=[text_input, encoder_input],
                    outputs=[entity_pred])
```

```python
import tensorflow as tf
Entity_model.compile(
  optimizer=tf.keras.optimizers.Adam(),
  loss='categorical_crossentropy',
  metrics=['acc'])
```

```python
Entity_model.fit({'html-text': Etrain_txt, 'img': Etrain_img},
          {'is_entity': Etrain_lab},
          epochs=5,
          batch_size=32)
```

```python
Entity_model.evaluate({'html-text': Etest_txt, 'img': Etest_img},
          {'is_entity': Etest_lab})
```

```python
m = tf.keras.metrics.AUC(num_thresholds=10)
predY = Entity_model.predict({'html-text': Etest_txt, 'img': Etest_img})
m.update_state(Etest_lab,predY)
print('AUC result for entity: ', m.result().numpy()) 
```

```python
categories = []
category_data = []
with open("./more_label.csv") as f:
    csv_reader = csv.DictReader(f)
    for row in csv_reader:
        if row['NewCategory'] != "":
            if not row['NewCategory'] in categories:
                categories.append(row['NewCategory'])
            category_data.append(row)
random.shuffle(category_data)
```

```python

Ctxt = []
Cimg = []
Clab = []
for row in category_data:
    Ctxt.append(textseq[LocTxt[row['html'].split('/')[-1]]])
    Cimg.append(img[LocImg[row['render'].split('/')[-1]]])
    lab = (row["NewCategory"]==np.array(categories)).astype('float')
    Clab.append(lab)
Ctxt = np.array(Ctxt)
Cimg = np.array(Cimg)
Clab = np.array(Clab)
```

```python
partition_rate = 0.9
cnt = int(len(Ctxt)*partition_rate)
Ctrain_txt = Ctxt[:cnt]
Ctrain_img = Cimg[:cnt]
Ctrain_lab = Clab[:cnt]
Ctest_txt = Ctxt[cnt:]
Ctest_img = Cimg[cnt:]
Ctest_lab = Clab[cnt:]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

for i in range(10):
    image = Cimg[i]
    label = Clab[i]
    
    print("Image shape: ", image.shape)
    print("Label: ", categories[np.argmax(label)])
    image_ = np.squeeze(image) # you can give axis attribute if you wanna squeeze in specific dimension
    plt.imshow(image_)
    plt.show()
    
```

```python

text_input2 = keras.Input(shape=(None,), name='html-text')  # Variable-length sequence of ints
encoder_input2 = keras.Input(shape=(224, 224, 3), name='img')

# Embed each word in the text into a 64-dimensional vector
text_features2 = layers.Embedding(num_words, 64)(text_input2)
# Reduce sequence of embedded words in the title into a single 128-dimensional vector
text_features2 = layers.LSTM(128)(text_features2)

# Use pretrained model for img feature extraction
feature_extractor_layer2 = hub.KerasLayer(feature_extractor_url,
                                         input_shape=(IMG_HEIGHT,IMG_WIDTH,3))
feature_extractor_layer2.trainable = False
img_features2 = feature_extractor_layer2(encoder_input2)

yyy = layers.concatenate([text_features2, img_features2])
category_pred = layers.Dense(num_categories, activation='softmax', name='category')(yyy)
#model = keras.Model(inputs=[text_input, encoder_input],
#                    outputs=[entity_pred, category_pred])
Category_model = keras.Model(inputs=[text_input2, encoder_input2],
                    outputs=[category_pred])
```

```python
Category_model.compile(
  optimizer=tf.keras.optimizers.Adam(),
  loss='categorical_crossentropy',
  metrics=['acc'])
Category_model.fit({'html-text': Ctrain_txt, 'img': Ctrain_img},
          {'category': Ctrain_lab},
          epochs=5,
          batch_size=32)
```

```python
Category_model.evaluate({'html-text': Ctest_txt, 'img': Ctest_img},
          {'category': Ctest_lab})
```
