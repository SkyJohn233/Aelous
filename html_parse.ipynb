```python
class HtmlHandler(SGMLParser):
    def reset(self):
        self.inTitle = False
        self.inPara = False
        self.result = ""
        self.ret = ""
        SGMLParser.reset(self)

    def start_title(self, attrs):
        self.inTitle = True
        
    def end_title(self):
        self.inTitle = False
        
    def start_p(self, attrs):
        self.inPara = True
        
    def end_p(self):
        self.inPara = False
        
    def handle_data(self, text):
        pflag = False
        if self.inTitle or self.inPara:
            for c in text:
                if c != ' ' and c != '' and c != '\n' and c != '\t':
                    pflag = True
                    break
            if pflag:
                self.result += text
                self.result += " "
            
    def ret_result(self):
        for c in self.result:
            if c != '\n' and c != '\t':
                self.ret += c
        return self.ret
```
